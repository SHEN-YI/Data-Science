```python
import pandas as pd
from dateutil.parser import parse
from datetime import datetime
moviedata=pd.read_csv('merged.csv')
moviedata=moviedata.loc[moviedata['Runtime'] != 'Runtime']
```

```python
moviedata
```

```python
moviedata.columns.values
```

# Select the time from 2000 to 2017

```python
factorslist=['Title','Year','Theaters','Country','Genre','Rated','Runtime','Open','Director','Actors','Language','imdbRating','Total Gross']
```

```python
moviepredictfactors=moviedata[factorslist]
```

```python
moviepredictfactors=moviepredictfactors[moviepredictfactors.Year.str.contains("–") == False]
```

```python
moviepredictfactors['Year']=moviepredictfactors['Year'].astype(int)
```

```python
moviepredictfactors_2000=moviepredictfactors[moviepredictfactors['Year'] >= 2000]
```

```python
moviepredictfactors_2000_2017=moviepredictfactors_2000[moviepredictfactors_2000['Year'] <= 2017]
```

```python
moviedata=moviepredictfactors_2000_2017
```

# Dealing with Theaters
## Fill the n/a with 0 

```python
moviedata['Theaters'].value_counts()
```

```python
moviedata.Theaters=moviedata.Theaters.iloc[:].apply(lambda x: x.split("'")[1].replace(',','').replace('n/a','0')).astype(int)
```

```python
moviedata.loc[:,'Theaters'].value_counts()
```

# Dealing with Country, Genre, Rated, Runtime, Director, Actors,Language

```python
moviedata
```

```python
moviedata.Country=moviedata.Country.astype(str)
moviedata.Country=moviedata.Country.iloc[:].apply(lambda x: x.split(",")[0])
```

```python
moviedata.Genre=moviedata.Genre.astype(str)
moviedata.Genre=moviedata.Genre.iloc[:].apply(lambda x: x.split(",")[0])
```

```python
moviedata.Rated.value_counts()
```

```python
moviedata.Rated[moviedata['Rated'] == 'Not Rated']='Unrated'
moviedata.Rated[moviedata['Rated'] == 'NOT RATED']='Unrated'
moviedata.Rated[moviedata['Rated'] == 'UNRATED']='Unrated'
```

```python
moviedata.Runtime[moviedata['Title']== 'El sueño del regreso']='100 min'
moviedata.Runtime[moviedata['Title']== 'Greg & Gentillon']='85 min'
moviedata.Runtime=moviedata.Runtime.astype(str)
moviedata.Runtime=moviedata.Runtime.iloc[:].apply(lambda x: x.split("min")[0].replace('nan','0'))
moviedata.Runtime=moviedata.Runtime.astype(int)
```

```python
moviedata
```

```python
moviedata.Director=moviedata.Director.astype(str)
moviedata.Director=moviedata.Director.iloc[:].apply(lambda x: x.split(",")[0])
```

```python
moviedata.Language=moviedata.Language.astype(str)
moviedata.Language=moviedata.Language.iloc[:].apply(lambda x: x.split(",")[0])
```

```python
moviedata.Actors=moviedata.Actors.astype(str)
moviedata.Actors=moviedata.Actors.iloc[:].apply(lambda x: x.split(",")[0:3])
```

```python
moviedata.imdbRating=moviedata.imdbRating.astype(float)
```

```python
moviedata['Total Gross']=moviedata['Total Gross'].astype(str)
moviedata['Total Gross']=moviedata['Total Gross'].iloc[:].apply(lambda x: x.split("'")[1].replace('$','').replace(',','').replace('n/a','0').replace('*','0'))
moviedata['Total Gross']=moviedata['Total Gross'].astype(float)
```

```python
moviedata.Open[moviedata['Open']== '2018']='1/1/2018'
moviedata.Open=moviedata.Open.astype(str)
moviedata.Open=moviedata.Open.iloc[:].apply(lambda x: x.split("b")[1].replace('TBD','1/1/2018').replace('N/A','1/1/2018').replace("'",''))
moviedata.Open.value_counts()
```

```python
try:
    moviedata.Open.iloc[:].apply(lambda x: datetime.strptime(x,'%m/%d/%Y'))
except:
    pass
```
