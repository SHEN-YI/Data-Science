```python
import numpy as np
res=[]
for actor in moviedata['Actors']:
    record=[]
    actor=actor.replace("[",'').replace("]",'').replace("'",'')
    actors=actor.split(',')
    for i in actors:
        val=actordata['Average Gross for each movie'][actordata['Person'] == i.strip()]
        if val.isnull().all():
            pass
        else:
            record.append(val.iloc[0])
    if not record:
        averes=0
    else:
        averes=max(record)
    print(averes)
    res.append(averes)
```
